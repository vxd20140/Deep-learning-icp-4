```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('data.csv')
df.head()
df.describe()
df.fillna(df.mean(), inplace=True)
print(df)
df[['Calories', 'Maxpulse']].agg(['min','max','count','mean'])
df[(df['Calories'] >= 500) & (df['Calories'] <= 1000)]
df[(df['Calories'] > 500) & (df['Maxpulse'] < 100)]
df = df.drop('Maxpulse', axis=1)
print(df)
df['Calories'] = df['Calories'].astype(int)
plt.scatter(df['Duration'], df['Calories'])
plt.xlabel("Duration")
plt.ylabel("Calories")
plt.show()
```

```python
import pandas as pd 
salariesData = pd.read_csv('Salary_Data.csv')
X = salariesData.iloc[:, :-1].values
Y= salariesData.iloc[:, 1].values
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 1/3, random_state = 0)
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
pred = regressor.predict(X_test)
from sklearn.metrics import mean_squared_error
mse1=mean_squared_error(X_test,pred)
mse = mean_squared_error(Y_test, pred)
import matplotlib.pyplot as plt
plt.scatter(X_train, Y_train, color = 'blue')
plt.scatter(X_test, Y_test, color = 'red')
plt.title('Salary Data')
plt.xlabel('Experience (Years)')
plt.ylabel('Salary')
plt.show()
```
